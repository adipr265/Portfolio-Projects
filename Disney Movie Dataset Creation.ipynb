```python
def in_case_value_list(row_data):
    if row_data.find('li'):
        return [li.get_text(' ', strip=True).replace('\xa0',' ') for li in row_data.find_all('li')] 
    elif row_data.find('br'):
        return[text for text in row_data.stripped_strings]
    else:
        return row_data.get_text(' ', strip=True).replace('\xa0',' ')

def clean_tags(soup1):
    for tags in soup1.find_all(['sup','span']):
        tags.decompose()

def get_url(website):
    
    url = requests.get(website)

    soup1 = BeautifulSoup(url.content)

    info_box = soup1.find(class_='infobox vevent')
    table_rows = info_box.find_all("tr")
    
    clean_tags(soup1)

    movie_dict = {}
    
    for index, row in enumerate(table_rows):
        if index == 0:
            movie_dict['title'] = row.find('th').get_text(' ', strip=True)  
        else:
            header = row.find('th')
            if header:
                movie_dict_key = row.find("th").get_text(' ', strip=True)
                movie_dict_value = in_case_value_list(row.find("td"))
                movie_dict[movie_dict_key] = movie_dict_value

    return(movie_dict)
```

```python
import re

amounts = r"thousand|million|billion"
number = r"\d+(,\d(3))*\.*\d*"

word_re = rf"\${number}(-|\sto\s)?({number})?\s({amounts})"
number_re = rf"\${number}"

def word_to_value(word):
    word_dict = {'thousand': 1000, 'million': 1000000, 'billion':1000000000, 'Million':1000000}
    return word_dict[word]

def parse_word_syntax(string):
    amt = re.search(number, string).group()
    value = float(amt.replace(",",""))
    word = re.search(amounts, string, flags=re.I)
    word_value = word_to_value(word.group())
    return value*word_value

def parse_value_syntax(string):
    amt = re.search(number, string).group()
    value = float(amt.replace(",",""))
    return value

def money_conversion(money):
    
    if money == 'N/A':
        return None
    
    if isinstance(money, list):
        money = money[0]
    
    word_syntax = re.search(word_re, money, flags=re.I)
    value_syntax = re.search(number_re, money)
    
    if word_syntax:
        return parse_word_syntax(word_syntax.group())
    
    elif value_syntax:
        return parse_value_syntax(value_syntax.group())
    
    else:
        return None
```

```python
for movie in movie_list:
    movie['Budget (float)'] = money_conversion(movie.get('Budget', 'N/A'))
    movie['Box office (float)'] = money_conversion(movie.get('Box office', 'N/A'))
```

```python
from datetime import datetime

dates = [movie.get('Release date', 'N/A') for movie in movie_list]

def clean_date(date):
    return date.split('(')[0].strip()
    
def date_conversion(date):
    if isinstance(date, list):
        date = date[0]
        
    if date == 'N/A':
        return None
    
    date_str = clean_date(date)
    
    fmts = ['%B %d, %Y', '%d %B %Y']
    for fmt in fmts:
        try:
            return datetime.strptime(date_str, fmt)
        except:
            pass
    return None
    
for date in dates:
    print(date_conversion(date))
    print()
```

```python
for movie in movie_list:
    movie['Release date (datetime)'] = date_conversion(movie.get('Release date', 'N/A'))
```

```python
import requests
import urllib

def get_rating_info(title):
    base_url = 'http://www.omdbapi.com/?'
    parameters = {'apikey': 'dc924049', 't': title}
    p_encoded = urllib.parse.urlencode(parameters)
    full_url = base_url + p_encoded
    return requests.get(full_url).json()

def get_rotten_tomato_score(rotten_rating):
    ratings = rotten_rating.get('Ratings', [])
    for rating in ratings:
        if rating['Source'] == 'Rotten Tomatoes':
            return rating['Value']

ro = get_rating_info('into the woods')
get_rotten_tomato_score(ro)
```
